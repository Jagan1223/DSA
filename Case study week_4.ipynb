# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
#count the values in each type of Buyer Gender and plot bar graph on it
ax=data['Buyer Gender'].value_counts().plot(kind='bar',figsize=(10,7),color=['blue','red'],fontsize=13)

ax.set_title('Male vs Female buyers',fontsize=15)
ax.set_xlabel('Gender',fontsize=15)
ax.set_ylabel('Count',fontsize=15)

#adding the count of each type in the graph
for i in ax.patches:
    ax.text(i.get_x()+.15,i.get_height()-300,str(i.get_height()),fontsize=22,color='white',fontweight='bold')
```

Female buyers are more than male buyers in the given data by 104 buyers
