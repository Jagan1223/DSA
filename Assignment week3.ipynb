# 1. Read the dataset to python environment.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('iris.xls')
data
```

# 2. Display the columns in the dataset.

```python
data.columns
```

# 3. Calculate the mean of each column of the dataset.

```python
print('Mean of SL is: ',round(data['SL'].mean(),2))
print('Mean of SW is: ',round(data['SW'].mean(),2))
print('Mean of PL is: ',round(data['PL'].mean(),2))
print('Mean of PW is: ',round(data['PW'].mean(),2))
```

# 4. Check for the null values present in the dataset.

```python
if((data.isna().values.any())==True):
    print('Null values present')
else:
    print('No null values present')
```

# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

### Count-Plot

```python
sns.countplot(x='Classification',data=data)
plt.grid()
```

From the above plot it is clear that the dataset contains three classifications:- Iris-setosa, Iris-versicolor and Iris-virginica, and the three classifications are of 50 each.

### Pair-Plot

```python
sns.pairplot(data)
```

From the above different combinations we can see that the PL and PW is having a relationship, that is the PL increases as the PW increases

### Box-Plot

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=data)
plt.grid()

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=data)
plt.grid()

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=data)
plt.grid()

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=data)
plt.grid()

plt.tight_layout()
```

From the boxplots we can have the following insights:-
   - Most of the Iris-setosa is having lower sepal length and higher sepal width
   - Iris-setosa is having the smallest petals, whereas Iris-virginica is having the largest petals
   - Sepals and petals of Iris-versicolor is having moderate size
